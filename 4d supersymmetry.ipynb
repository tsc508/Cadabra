We consider 4d N=1 supersymmetry, and use Cadabra to calculate the supersymmetric transformations of component fields, with the help of superfield. 

```cadabra2
\tbar{#}::LaTeXForm("\bar\theta").
\cbar{#}::LaTeXForm("\bar\chi").
\lambar{#}::LaTeXForm("\bar\lambda").
\sigbar{#}::LaTeXForm("\bar\sigma").
\aldot{#}::LaTeXForm("\dot\alpha").
\betdot{#}::LaTeXForm("\dot\beta").
\gamdot{#}::LaTeXForm("\dot\gamma").
\zetdot{#}::LaTeXForm("\dot\zeta").
\etdot{#}::LaTeXForm("\dot\eta").
{\alpha,\beta,\gamma,\zeta,\eta}::Indices("dotted",position=independent,values={1,2});
{\aldot,\betdot,\gamdot,\zetdot,\etdot}::Indices("undotted",position=independent,values={1,2});
{\mu,\nu,\rho,\sigma}::Indices("vector",position=fixed,values={0,1,2,3});
\sigma^{\mu}::ImplicitIndex(\sigma^\mu_{\alpha \aldot});
\sigma_{\mu}::ImplicitIndex(\sigma_{\mu \alpha \aldot});
\sigbar^{\mu}::ImplicitIndex(\sigbar^\mu_{\alpha \aldot});
\sigbar_{\mu}::ImplicitIndex(\sigbar_{\mu \alpha \aldot});
\theta::ImplicitIndex(\theta^{\alpha});
\tbar::ImplicitIndex(\tbar^{\aldot});
explicit_indices($(\theta\theta)(\theta \sigma^\mu \tbar)$);
```

First we define a general superfield $F$. Since PartialDerivative in Cadabra can only be applied to bosonic terms, we have to write the expression in a carefully designed way, otherwise the Leibniz law fails. \
\begin{equation}
F(x,\theta,\bar\theta)=f(x)+\theta\phi(x)+\bar\theta\bar\chi(x)\\+\theta\theta m(x)+\bar\theta\bar\theta n(x)+\theta\sigma^m\bar\theta v_m(x)\\+\theta\theta\bar\theta\bar\lambda(x)+\bar\theta\bar\theta\theta\psi(x)+\theta\theta\bar\theta\bar\theta d(x)
\end{equation}

```cadabra2
Superfield:=F->(f+\theta^{\alpha} \phi_{\alpha}+\tbar_{\aldot} \cbar^{\aldot}
+\indexbracket{\theta \theta} m+\indexbracket{\tbar \tbar} n+\indexbracket{\theta \sigma^\mu \tbar} v_{\mu} 
+\indexbracket{\theta \theta} \tbar_{\aldot} \lambar^{\aldot})+\indexbracket{\tbar \tbar} \theta^{\alpha} \psi_{\alpha}
+\indexbracket{\theta \theta} \indexbracket{\tbar \tbar} d);
Testfield:=F->\indexbracket(\theta \sigma^\mu \tbar)v_\mu;
```

Attach corresponding properties to partial derivatives in the superspace. $\partial,\bar\partial$ act on $\theta,\bar\theta$, while $\hat\partial$ acts on $x^\mu$.

```cadabra2
\parbar{#}::LaTeXForm("\bar\partial").
\partialmu{#}::LaTeXForm("\hat\partial").
\xibar{#}::LaTeXForm("\bar\xi").
\partial{#}::ImplicitIndex(\partial{#}_{\theta^{#}});
\parbar{#}::ImplicitIndex(\parbar{#}_{\tbar^{#});
{\partial{#},\parbar{#},\partialmu{#}}::PartialDerivative;
```

```cadabra2
\epsilon_{\alpha \beta}::AntiSymmetric;
\epsilon^{\alpha \beta}::AntiSymmetric;
\epsilon_{\aldot \betdot}::AntiSymmetric;
\epsilon^{\aldot \betdot}::AntiSymmetric;
{\delta_{\alpha}^{\beta},\delta^{\alpha}_{\beta}}::KroneckerDelta;
{\delta^{\aldot}_{\betdot},\delta_{\aldot}^{\betdot}}::KroneckerDelta;
```

The operator generating an infinitesimal susy transformation of the superfield is defined as\
\begin{equation}
	\xi Q+\bar\xi \bar Q=\xi^\alpha(\frac{\partial}{\partial\theta^{\alpha}}-i(\sigma^\mu)_{\alpha\dot\alpha}\bar\theta^{\dot\alpha}\partial_\mu)+\bar\xi_{\dot\alpha}(\frac{\partial}{\partial\bar\theta_{\dot\alpha}}-i\theta^\alpha(\sigma^\mu)_{\alpha\dot\beta}\epsilon^{\dot\beta\dot\alpha}\partial_\mu)
\end{equation}

```cadabra2
susy:=\xi^{\alpha} \partial_{\alpha}{F}-i \xi^{\alpha} \sigma^{\mu}_{\alpha \aldot} \tbar^{\aldot} \partialmu_{\mu}{F}
+\xibar_{\aldot} \parbar^{\aldot}{F}+i \theta^{\alpha} \sigma^{\mu}_{\alpha\aldot} \xibar^{\aldot} \partialmu_{\mu}{F};
```

we specify how the derivatives act on the superfields, and whom they act on,

```cadabra2
superderivative:={
\partial_{\alpha}{\theta^{\beta}}->\delta_{\alpha}^{\beta},
\parbar^{\aldot}{\tbar_{\betdot}}->\delta^{\aldot}_{\betdot},
\partial_{\alpha}{\theta_{\beta}}->\epsilon_{\beta \alpha},
\parbar^{\aldot}{\tbar^{\betdot}}->\epsilon^{\betdot \aldot},
\partial_{\alpha}{\indexbracket(\theta \theta)}->2\theta_{\alpha},
\parbar^{\aldot}{\indexbracket{\tbar \tbar}}->2\tbar^{\aldot},
\partial_{\alpha}{\indexbracket{\theta \sigma^{\mu} \tbar}}->\sigma^\mu_{\alpha \aldot} \tbar^{\aldot},
\parbar^{\aldot}{\indexbracket{\theta \sigma^{\mu} \tbar}}->-\theta^{\alpha}\sigma^\mu_{\alpha \betdot}\epsilon^{\betdot \aldot}
};
{\theta^{#},\indexbracket{\theta \theta}}::Depends(\partial{#});
{\tbar_{#},\indexbracket{\tbar \tbar}}::Depends(\parbar{#});
\indexbracket{\theta \sigma^\mu \tbar}::Depends(\partial{#},\parbar{#});
{f,\phi_{\alpha},\cbar^{\aldot},m,n,v_{\mu},\lambar^{\aldot},\psi_{\alpha},d}::Depends(\partialmu{#});
```

Now we perform the susy transformation on the superfield. Product_rule applies the Leibniz rule, while unwrap discards the irrelevant derivatives. 

```cadabra2
substitute(susy,Superfield);
distribute(susy)
product_rule(susy)
distribute(susy)
unwrap(susy)
substitute(susy,superderivative);
```

Now we have get rid of the Grassmann derivatives, and have more freedom to simplify. Before further calculation, first we modify our post_process function

```cadabra2
def post_process(ex):
	distribute(ex)
	product_rule(ex)
	unwrap(ex)
	eliminate_kronecker(ex)
	canonicalise(ex)
	collect_terms(ex)
```

and attach anticommuting property to the spinors

```cadabra2
{\theta^{\alpha},\tbar^{\aldot},\theta_{\alpha},\tbar_{\aldot},\xi^{\alpha},\xibar_{\aldot},\phi_{\alpha},\phi^{\alpha},\cbar_{\aldot},\cbar^{\aldot},\lambar_{\aldot},\lambar^{\aldot},\psi_{\alpha},\psi^{\alpha}}::AntiCommuting;
{\theta^{\alpha},\tbar^{\aldot},\theta_{\alpha},\tbar_{\aldot},\xi^{\alpha},\xi_{\alpha},\xibar^{\aldot},\xibar_{\aldot},\phi_{\alpha},\phi^{\alpha},\cbar_{\aldot},\cbar^{\aldot},\lambar_{\aldot},\lambar^{\aldot},\psi_{\alpha},\psi^{\alpha}}::SelfAntiCommuting;
```

next we teach Cadabra how to contract $\epsilon$ with the spinors, and how to contract two $\epsilon$:

```cadabra2
eps_contraction:={
\epsilon_{\alpha \beta} A?^{\beta}->A?_{\alpha},
\epsilon^{\alpha \beta} A?_{\beta}->A?^{\alpha},
\epsilon^{\aldot \betdot} A?_{\betdot}-> A?^{\aldot},
\epsilon_{\aldot \betdot} A?^{\betdot}-> A?_{\aldot},
-\epsilon^{\alpha \beta} \epsilon_{\beta \gamma}-> -\delta^{\alpha}_{\gamma},
-\epsilon^{\aldot \betdot} \epsilon_{\betdot \gamdot}-> -\delta^{\aldot}_{\gamdot},
,
};
substitute(susy, eps_contraction,repeat=True);
```

note that any term with 3 or more $\theta$ or $\bar\theta$ vanishes, because they are 2-component spinors, so we raise all the indices, then delete these terms:

```cadabra2
explicit_index:={
\indexbracket(\theta \theta)->\theta^{\alpha} \theta_{\alpha},
\indexbracket(\tbar \tbar)->\tbar_{\aldot} \tbar^{\aldot},
\indexbracket(\theta \sigma^\mu \tbar)->\theta^{\alpha} \sigma^{\mu}_{\alpha \aldot} \tbar^{\aldot},
};
raise_indices:={\theta_{\alpha}->\epsilon_{\alpha \beta}\theta^{\beta},\tbar_{\aldot}->\epsilon_{\aldot \betdot}\tbar^{\betdot}};
substitute(susy, explicit_index);
substitute(susy, raise_indices);
zoom(susy, ${\theta^{\alpha}\theta^{\beta}\theta^{\gamma},\tbar^{\aldot}\tbar^{\betdot}\tbar^{\gamdot}$);
substitute(susy, ${\theta^{\alpha}->0,\tbar^{\aldot}->0}$);
unzoom(susy)
substitute(susy, eps_contraction,repeat=True);
```

Next we find anything that can be written as $(\theta\theta)$ or $(\bar\theta \bar\theta)$. After raising the indices, we can adopt the formula\\
\begin{equation}
	\theta^\alpha\theta^\beta=-\frac{1}{2}\epsilon^{\alpha\beta}\theta\theta,
	\bar\theta^{\dot\alpha}\bar\theta^{\dot\beta}=\frac{1}{2}\epsilon^{\dot\alpha\dot\beta}\bar\theta\bar\theta,
\end{equation}

```cadabra2
tt_contraction:={
\theta^{\alpha}\theta^{\beta}|\alpha != \beta->-1/2 \epsilon^{\alpha \beta} \indexbracket{\theta\theta},
\tbar^{\aldot}\tbar^{\betdot}|a != b->1/2 \epsilon^{\aldot \betdot} \indexbracket{\tbar\tbar},

};
substitute(susy, eps_contraction,repeat=True)
substitute(susy, raise_indices,repeat=True)
substitute(susy, tt_contraction,repeat=True)
substitute(susy, eps_contraction,repeat=True);
```

Another type of combination is $\theta\bar\theta$, for which we use the formula\\
\begin{equation}
	\theta^{\alpha}\bar\theta^{\dot\alpha}=-\frac{1}{2}(\theta\sigma^\mu\bar\theta)\bar\sigma^{\dot\alpha \alpha}
\end{equation}

```cadabra2
Fierz_contraction:={
\theta^{\alpha}\tbar^{\aldot}->-1/2\indexbracket{\theta \sigma^\mu \tbar} \sigbar_{\mu}^{\aldot\alpha}
};
substitute(susy, raise_indices,repeat=True)
substitute(susy, Fierz_contraction)
substitute(susy, eps_contraction,repeat=True);
```

```cadabra2
general_contraction:={
i i->-1,
A?^{\alpha} B?_{\alpha}->\indexbracket(A? B?),
A?_{\aldot} B?^{\aldot}->\indexbracket(A? B?),
A?^{\alpha} \sigma^\mu_{\alpha \aldot} B?^{\aldot}->\indexbracket(A? \sigma^\mu B?)
};
substitute(susy, general_contraction)
substitute(susy, eps_contraction);
```

After the calculation, we recover post_process to its initial definition:

```cadabra2
def post_process(ex):
    collect_terms(ex)
```

```cadabra2
susy:=\xi^{\alpha} \partial_{\alpha}{F}-i \xi^{\alpha} \sigma^{\mu}_{\alpha \aldot} \tbar^{\aldot} \partialmu_{\mu}{F}
+\xibar_{\aldot} \parbar^{\aldot}{F}+i \theta^{\alpha} \sigma^{\mu}_{\alpha\aldot} \xibar^{\aldot} \partialmu_{\mu}{F}.
substitute(susy,Superfield)
distribute(susy)
product_rule(susy)
distribute(susy)
unwrap(susy)
substitute(susy,superderivative)
def post_process(ex):
	distribute(ex)
	product_rule(ex)
	unwrap(ex)
	eliminate_kronecker(ex)
	canonicalise(ex)
	collect_terms(ex)
substitute(susy, eps_contraction,repeat=True)
substitute(susy, explicit_index)
substitute(susy, raise_indices)
zoom(susy, ${\theta^{\alpha}\theta^{\beta}\theta^{\gamma},\tbar^{\aldot}\tbar^{\betdot}\tbar^{\gamdot}}$)
substitute(susy, ${\theta^{\alpha}->0,\tbar^{\aldot}->0}$)
unzoom(susy)
substitute(susy, eps_contraction,repeat=True)
substitute(susy, raise_indices,repeat=True)
substitute(susy, tt_contraction,repeat=True)
substitute(susy, eps_contraction,repeat=True)
substitute(susy, raise_indices,repeat=True)
substitute(susy, Fierz_contraction)
substitute(susy, eps_contraction,repeat=True)
substitute(susy, general_contraction)
substitute(susy, eps_contraction);
def post_process(ex):
    collect_terms(ex)
```
